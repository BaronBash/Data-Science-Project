```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
###Age against annual income
```

```python
X = df.iloc[:,[2,3]].values
```

```python
wss = []
for i in range (1,11):
    km = KMeans(n_clusters=i, init = 'k-means++', max_iter = 300, n_init=10, random_state =5)
    km.fit(X)
    wss.append(km.inertia_)
```

```python
plt.plot(range(1,11), wss)
plt.title('The Elbow Method')
plt.xlabel('range')
plt.ylabel('wss')
plt.show()
```

```python
km = KMeans(n_clusters =4, init='k-means++',max_iter = 300, n_init = 10, random_state=4)
y_means = km.fit_predict(X)
```

```python
y_means
```

```python
plt.style.use('ggplot')
plt.style.use('fivethirtyeight')
plt.figure(figsize= (8,8))
plt.scatter(X[y_means == 0,0], X[y_means == 0,1], s =100, c='red',label = 'Old below average income earners')
plt.scatter(X[y_means == 1,0], X[y_means == 1,1], s =100, c='yellow',label = 'Low income earners')
plt.scatter(X[y_means == 2,0], X[y_means == 2,1], s =100, c='blue',label = 'Middle aged High income earners')
plt.scatter(X[y_means == 3,0], X[y_means == 3,1], s =100, c='green',label = 'Young average income earners')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], s =100, c='black',label = 'datapoint')
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.title('Customer Grouping Based on Age and Annual income')
plt.legend()
plt.show()
```
